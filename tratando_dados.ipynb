## Limpeza de dados

```python
auxiliar = df.loc[13639]
auxiliar
```

```python
def limpa_preco(linha):
  linha['Price'] = linha['Price'].replace('R$ ', '')
  return linha
```

```python
limpa_preco(auxiliar)
```

```python
df = df.apply(lambda x: limpa_preco(x), axis=1)
df
```

```python
def ajusta_preco(linha):
  linha['Price'] = linha['Price'].replace('.', '')
  preco = int(linha['Price'])
  if(preco < 10000):
    preco = preco * 200
  linha['Price'] = preco
  return linha
```

```python
auxiliar = df.loc[13639]
auxiliar
```

```python
ajusta_preco(auxiliar)
```

```python
df = df.apply(lambda x: ajusta_preco(x), axis=1)
df
```
