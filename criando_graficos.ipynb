## Exibindo gráficos de filmes

Configurando eixo x e y do gráfico

```python
df = pd.read_csv('Best Movies Netflix.csv')
eixo_x = df['MAIN_GENRE'].value_counts().keys()
eixo_y = df['MAIN_GENRE'].value_counts().values
```

Configurando propriedades do gráfico

```python
plt.bar(eixo_x, eixo_y)
plt.title('Filmes por genero')
plt.xticks(rotation=90)
plt.xlabel('Gênero de filmes')
plt.ylabel('Quantidade de filmes')
plt.show()
```

```python
eixo_x = df['RELEASE_YEAR'].value_counts().sort_index().keys()
eixo_y = df['RELEASE_YEAR'].value_counts().sort_index().values
```

```python
plt.plot(eixo_x, eixo_y)
plt.title('Filmes por ano de lançamento')
plt.xlabel('Ano de lançamento')
plt.ylabel('Quantidade de filmes')
plt.show()
```
