# Analisando dados

```python
import pandas as pd
import seaborn as sns
```

## Analisando dados de viagens de taxi em Nova York

```python
taxis = sns.load_dataset("taxis")
taxis.head()
```

```python
taxis['dropoff_borough'].value_counts()
```

## Hipotese: Vizinhanças em que as casas costumam ser mais caras, oferecem melhores gorjetas?

```python
taxis.groupby(['dropoff_borough'])['tip'].agg(['mean', 'median', 'max', 'min'])
```

- Com estes dados a hipotese é refutada, porém com mais dados pode ser revista esta hipotese

### Dados gorjeta recebidas por um garçom
Hipoteste: Clientes fumantes oferecem gorjetas mais altas do que os não fumantes

```python
data_set = sns.load_dataset("tips")
data_set.head()
```

```python
data_set['smoker'].value_counts()
```

```python
data_set.groupby(['smoker'])['tip'].agg(['mean', 'median', 'max', 'min'])
```

Resposta: seguindo a análise, os clientes fumantes realmente dão gorjetas mais altas do que clientes não fumantes

Hipotese: Finais de semana (Sábado e Domingo) costumam ter gorjetas mais altas

```python
data_set['day'].value_counts()
```

```python
data_set.groupby(['day'])['tip'].agg(['mean', 'median', 'max', 'min'])
```

Resposta: Analisando os dados, percebe-se que sábado e domingo os clientes dão gorjetas mais altas
