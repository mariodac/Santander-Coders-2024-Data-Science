# Buscando correlações entre os dados

```python
import pandas as pd
import seaborn as sns
```

Obtendo lista de datasets do seaborn

```python
sns.get_dataset_names()
```

Utilizando dataset iris, dados de especies da flor iris

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris, hue = 'species')
```

## Exercicio: Correlação de dados do dataset titanic

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic['class'].value_counts()
```

```python
sns.pairplot(titanic, hue = 'class')
```

Utilizando 'pclass', observando o gráfico é possível a correlação com vários dados: fare(Tarifa do passageiro), age(idiade), sibsp(numero de parentes a bordo), parch(numero de pais a bordo), survived(sobreviveu)
